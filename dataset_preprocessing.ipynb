### Extracting the Genres of the respective films and the Keywords of the respective films and putting them into new columns

```python
import pandas as pd

# Function to extract and join genre names
def extract_genre_names(genre_list):
    genre_list = eval(genre_list)
    # Extracting the 'name' key from each dictionary in the list
    genre_names = [genre['name'] for genre in genre_list]
    # Joining the genre names into a single string, separated by commas
    return ', '.join(genre_names)

# Applying the function to the 'genre' column and assigning the result to a new column
tmdb['Genre Names'] = tmdb['genres'].apply(extract_genre_names)

# Display the first few rows to verify the new column
tmdb.head()
```

```python
# Function to extract and join keyword names
def extract_keyword_names(keyword_list):
    keyword_list = eval(keyword_list)
    # Extracting the 'name' key from each dictionary in the list
    keyword_names = [keyword['name'] for keyword in keyword_list]
    # Joining the keyword names into a single string, separated by commas
    return ', '.join(keyword_names)

# Applying the function to the 'keywords' column and assigning the result to a new column
tmdb['Keyword Names'] = tmdb['keywords'].apply(extract_keyword_names)

# Display the first few rows to verify the new column
tmdb.head()
```
